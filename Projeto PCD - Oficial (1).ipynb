```python
def gaussian(x, a, x0, sigma):
    '''Uma função que a partir dos parâmetros definidos gera a gaussiana'''
    return a * np.exp(-(x - x0)**2 / (2 * sigma**2))

def multi_gaussian(x, *params):
    '''geração de várias gaussianas a partir de vários parâmetros'''
    n = len(params) // 3
    y = np.zeros_like(x)
    for i in range(n):
        a = params[3*i]
        x0 = params[3*i+1]
        sigma = params[3*i+2]
        y += gaussian(x, a, x0, sigma)
    return y

def multi_gaussian_parametros(x, *params):
    '''Associa a cada gaussiana o seu parâmetro'''
    indice = 1
    n = len(params) // 3
    y = np.zeros_like(x)
    for i in range(n):
        a = params[3*i]
        x0 = params[3*i+1]
        sigma = params[3*i+2]
        y += gaussian(x, a, x0, sigma)
        
        print(f"Parametros da gaussiana {indice}: altura máxima {indice} = {round(a,3)} em abs, comprimento de onda do pico {indice} = {round(x0,3)} nm, FWHM {indice} = {round(sigma,3)}")
        indice += 1

def achar_picos_chato (file_pattern, x_min = 0 , x_max = float("inf"), h = 0, dist = None):
    """Faz uma tabela com os picos de absorbância com seus respectivos comproimentos de onda. Não plota um gráfico sinalizando esses picos!"""
    arquivos = gb.glob(file_pattern)
    indice = 1
    for i in arquivos: 
        dados = pd.read_csv(i)
        dados.rename(columns = {dados.columns[0]: "Comprimento de onda (nm)", dados.columns[1]: "Absorbância"}, inplace = True)
        
        comprimento_de_onda = dados["Comprimento de onda (nm)"].values
        
        dados_filtrados = dados[(dados["Comprimento de onda (nm)"] >= x_min) & (dados["Comprimento de onda (nm)"] <= x_max)]
        
        comprimento_de_onda = dados_filtrados["Comprimento de onda (nm)"].values
        absorbancia = dados_filtrados["Absorbância"].values
 
    prom = None

    p, _ = fp(x = dados_filtrados.Absorbância,
              height = h,
              prominence = prom,
              distance = dist
              )
   
    return dados_filtrados.iloc[p]

def lista_picos (file_pattern, x_min, x_max, h, dist):
    '''Gera uma lista com os picos com base na função acha picos'''
    h = 0.3
    a = achar_picos_chato (file_pattern, x_min, x_max, h, dist)
    c = a.values.tolist()
    known_peaks = []
    for i,v in enumerate(c):
        known_peaks.append (c[i][0])
    return known_peaks
    


def plt_gaussianas():
    '''Plota o gráfico da função'''
    arquivos = []
    apenas_um_arquivo = False
    
    while apenas_um_arquivo == False:
        file_pattern = input("Coloque o diretório e o padrão de arquivo desejado: (A função gaussiana comporta apenas um arquivo!) ")
        arquivos = gb.glob(file_pattern)
        if len(arquivos) == 1:
            file_pattern = arquivos[0]
            apenas_um_arquivo = True
        else:
            print("Quantidade inválida de arquivos")
            arquivos = []

            
#Ajustando o nome das colunas do arquivo
    data = pd.read_csv(file_pattern)

    data.rename(columns = {data.columns[0]: "Comprimento de onda (nm)", data.columns[1]: "Absorbância"}, inplace = True)

    wavelengths = data["Comprimento de onda (nm)"]

    absorbance = data["Absorbância"]
        
            
#Definindo os intervalos por input:
    determinar_intervalos = input("Você deseja determinar os intervalos de comprimento de onda? (S/N): ").strip().lower()
    det1 = True
    det2 = True
    while det1 == True or det2==True:
        if determinar_intervalos == "s":
            while det1==True:
                comprimento_de_onda_min = input("Digite um valor inteiro de comprimento de onda mínimo para os picos (ou pressione enter para não determinar esse dado): ")
                if comprimento_de_onda_min == "":
                    x_min = None
                    det1 = False
                elif comprimento_de_onda_min.isdigit():
                    x_min = int(comprimento_de_onda_min)
                    det1 = False
                else:
                    print("Entrada Inválida! Tente novamente.")
            while det2 == True:
                comprimento_de_onda_max = input("Digite um valor inteiro de comprimento de onda máximo para os picos(ou pressione enter para não determinar esse dado): ")
                if comprimento_de_onda_max == "":
                    x_max = None
                    det2 = False
                elif comprimento_de_onda_max.isdigit():
                    x_max = int(comprimento_de_onda_max)
                    det2= False
                else:
                    print("Entrada Inválida! Tente novamente.")
        elif determinar_intervalos == "n":
            x_min = 0
            x_max = float("inf")
            break
        else:
            print("Entrada Inválida! Tente novamente.")
            
        
#Definindo os parametros por input:
    determinar_parametros = input("Você deseja determinar os parametros de altura e distancia mínima dos picos? (S/N): ").strip().lower()
    det3 = True
    det4 = True
    h = None
    while det3 == True or det4 == True:
        if determinar_parametros == "s":
            while det3 == True:
                altura_min = input("Digite a altura mínima do pico (ou pressione enter para não determinar esse dado): ")
                if altura_min == "":
                    h = 0
                    det3 = False
                elif det3 == True:
                    try:
                        float(altura_min)
                        b = True
                        det3 = False
                    except ValueError:
                        b = False
                    if b == True:
                        h = float(altura_min)
                        det3 = False
                else:
                    print("Entrada Inválida! Tente novamente.")
            while det4 == True:
                distancia_picos = input("Digite um valor inteiro para a distância mínima entre dois picos (ou pressione enter para não determinar esse dado): ")
                if distancia_picos == "":
                    dist = None
                    det4 = False
                elif distancia_picos.isdigit():
                    dist = int(distancia_picos)
                    det4= False
                else:
                    print("Entrada Inválida! Tente novamente.")
        elif determinar_parametros == "n":
            h = 0
            dist = None
            break
        else:
            print("Entrada Inválida! Tente novamente.")

    # Título
    titulo = input("Digite o título do seu gráfico: (Dica: Seja informativo, mas conciso!): ")
    title = titulo  
    
    known_peaks=lista_picos(file_pattern,x_min,x_max,h,dist)
# Suposição inicial para os parâmetros da gaussianas: [a1, x0_1, sigma1, a2, x0_2, sigma2, ...]

    initial_guesses = []
    for peak in known_peaks:
        a_guess = max(absorbance)  # Initial guess for amplitude
        sigma_guess = 20  # Initial guess for width
        initial_guesses.extend([a_guess, peak, sigma_guess])

    popt, _ = curve_fit(multi_gaussian, wavelengths, absorbance, p0=initial_guesses)

    plt.plot(wavelengths, absorbance, label='Data')

# Plot da função multi-Gaussiana ajustada
    fitted_absorbance = multi_gaussian(wavelengths, *popt)
    multi_gaussian_parametros(wavelengths, *popt)
    
    plt.plot(wavelengths, fitted_absorbance, label='Fitted Multi-Gaussian', linestyle='--')

# Plot de cada individual comportamento da gaussiana
    for i in range(len(popt) // 3):
        a = popt[3*i]
        x0 = popt[3*i+1]
        sigma = popt[3*i+2]
        plt.plot(wavelengths, gaussian(wavelengths, a, x0, sigma), linestyle=':', label=f'Gaussian {i+1}')

    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Absorbance')
    plt.title(f"{title} - Deconvolução gaussiana")
    
    plt.legend()

    plt.savefig(f"{title} - Deconvolução gaussiana.png")
    plt.show()
```
