```python
def achar_picos (file_pattern,x_min = 0, x_max = float ('inf'), h = 0, dist = None, titulo = 'achar picos'):
    """Faz uma tabela com os picos de absorbância com seus respectivos comproimentos de onda. 
    Também plota um gráfico sinalizando esses picos.
    Ambos gráficos e tabelas são salvos como arquivos no diretório presente
    
    O único argumento essencial é um padrão de arquivo.
    Se desejar, o usuário também pode escolher as as seguintes variáveis: intervalo (x_min e x_max) - números inteiros, 
    altura mínima dos picos (h) - float, distância mínima considerada entre dois picos (d) - número inteiro e um título.
    
    """
    

    arquivos = gb.glob(file_pattern)


    #Função para encontrar picos
    
    arquivos = gb.glob(file_pattern)
    indice = 1
    print (arquivos)
    for i in arquivos: 
        dados = pd.read_csv(i)
        dados.rename(columns = {dados.columns[0]: "Comprimento de onda (nm)", dados.columns[1]: "Absorbância"}, inplace = True)
        
        comprimento_de_onda = dados["Comprimento de onda (nm)"].values
        
        dados_filtrados = dados[(dados["Comprimento de onda (nm)"] >= x_min) & (dados["Comprimento de onda (nm)"] <= x_max)]
        
        comprimento_de_onda = dados_filtrados["Comprimento de onda (nm)"].values
        absorbancia = dados_filtrados["Absorbância"].values
        prom = None

        p, _ = fp(x = dados_filtrados.Absorbância,
                  height = h,
                  prominence = prom,
                  distance = dist
                  )
        ax = dados_filtrados.plot ('Comprimento de onda (nm)', 'Absorbância')
        ax.set_title(f"{titulo} - Gráfico {indice}", fontsize=16, fontweight='bold')
        
        
        bx =sns.scatterplot (data =dados_filtrados.iloc[p].reset_index(),
                         x= 'Comprimento de onda (nm)',
                         y = 'Absorbância',
                         color = 'red', alpha = 0.5
                        )
        
        plt.savefig(f"{titulo} - Gráfico {indice}.png")
        plt.show()
        indice += 1
        
    print ('\n')
    
    #plotando e salvando a tabela com absorbância e comprimentos de ondas do pico
     df = dados_filtrados.iloc[p]
     novos_dados = pd.DataFrame (df)

     fig, ax = plt.subplots(figsize=(5, 2))  # Adjust the size as needed
     ax.axis('tight')
     ax.axis('off')
     table = ax.table(cellText=novos_dados.values, colLabels=novos_dados.columns, cellLoc='center', loc='center')


     plt.savefig('dataframe_image.png')
     plt.show()
```
